### Q.8: Steps in MLDLC(Machine learning development Life cycle)

**📌 All 11 steps:**
1. Peoblem Definition
- define what problem you want to solve

2. Data Collection
- Gather relevant data

3. Data Cleaning & Preprocessing
- Remove errors, handle missing values

4. Exploratory data Analysis(EDA)
- Understand data using graphs and statistics

5. Feature Engineering
- Create useful input  features

6. Model Selection
- choose the bast ML algorithm

7. Model Training 
- Train the model on data

8. model Evaluation 
- check performance (accuracy, etc.)

9. Hyperparameter Tuning 
- Improve model performance

10. Deployment 
- Use model in real-world application 

11. Monitoring & Maintenance
- Track and update model regularly

Problem → Data Collection → Data Cleaning → EDA → Feature Engineering  
→ Model Selection → Training → Evaluation → Tuning → Deployment → Monitoring
