### Q.1 5 examples of ML in Daily Life
In our daily life , we unknowingly interact with Machine Learning in many ways:

**1. Spem Email filtering**
Gmail automatically detects spam emails using ML algorithms.

**2. Netflix/YouTube Recommendations**
Suggests movies or videos based on our watch history.

**3. Google Maps Navigation**
Predicts traffic and suggests the fastest route.

**4. Voice Assistants (siri, Alexa, Google Assistant)**
Understands and responds to voice commands.

**5. Online shopping (Amazon/Flipkart)**
Recommends product based on browing and purchase history.

### Q.3: Differences Between ML and DL

| Aspect                  | Machine Learning (ML)             | Deep Learning (DL)                    |
| ----------------------- | --------------------------------- | ------------------------------------- |
| **Data Requirement**    | Works with small to medium data   | Requires large datasets               |
| **Computational Power** | Less power required               | High power (GPU/TPU needed)           |
| **Use Cases**           | Spam detection, prediction models | Image recognition, speech recognition |

### Q.5: ML Application Area – Healthcare

**Application Area**: Healthcare

**Problem Being Solved**:
Early detection of diseases like cancer and diabetes is difficult and time-consuming. ML helps in analyzing medical data quickly and accurately.

**ML Model Used**:
Common models include:

- Decision Trees
- Neural Networks
- Deep Learning (for image analysis like X-rays)

**Benefits**:
- Faster diagnosis
- Improved accuracy
- Reduces human error
- Saves lives through early detection
- Helps doctors make better decisions
