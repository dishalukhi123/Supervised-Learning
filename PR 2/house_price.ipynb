#### 1. What is Regularization in Machine Learning? Why is it needed?

Regularization is a technique used to reduce overfitting in machine learning models by adding a penalty term to the loss function. It helps control model complexity and improves generalization on unseen data. Regularization ensures the model does not memorize training data.


#### 2. Difference between Ridge Regression (L2) and Lasso Regression (L1)

| Ridge Regression (L2)      | Lasso Regression (L1)                  |
| -------------------------- | -------------------------------------- |
| Uses L2 penalty            | Uses L1 penalty                        |
| Reduces coefficient values | Can reduce coefficients to zero        |
| Keeps all features         | Performs feature selection             |
| Best for multicollinearity | Best for removing unimportant features |



#### 3. What is Cross-Validation and why is it important?

Cross-validation is a technique used to evaluate model performance by splitting data into multiple training and testing sets. It helps measure how well the model generalizes to unseen data. Cross-validation reduces overfitting and improves model reliability.


#### 4. Explain the following cross-validation techniques

**a) K-Fold Cross-Validation**

In K-Fold Cross-Validation, the dataset is divided into K equal parts. The model is trained on K−1 folds and tested on the remaining fold. This process repeats K times, and the average performance is calculated.


**b) Stratified K-Fold Cross-Validation**

Stratified K-Fold maintains the same class distribution in every fold. It is mainly used for classification problems with imbalanced datasets to ensure fair evaluation.


**c) Leave-One-Out Cross-Validation (LOOCV)**

LOOCV uses one data point as the testing set and the remaining data for training. This process repeats for every data point. It provides accurate evaluation but is computationally expensive.


**d) Time Series Split**

Time Series Split is used for sequential or time-dependent data. Training is performed on past data and testing on future data, preserving the time order of observations.


#### 5. Why are tree-based models less sensitive to feature scaling?

Tree-based models split data based on feature thresholds rather than distance calculations. Since they do not depend on feature magnitude, scaling techniques like StandardScaler or MinMaxScaler are usually unnecessary for decision trees and random forests.
