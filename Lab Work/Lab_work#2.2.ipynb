## 1. Define SST

### Meaning

SST (Total Sum of Squares) measures the total variation of y values around their mean.

### If SST = 0

All y values are identical.

There is no variation in the dataset.

---

## 2. Variance Decomposition

### Fundamental Identity

:contentReference[oaicite:1]{index=1}

Where:

### SSR (Regression Sum of Squares)

\[
SSR = \sum (\hat{y_i}-\bar{y})^2
\]

Measures variation explained by the regression model.

### SSE (Sum of Squared Errors)

\[
SSE = \sum (y_i-\hat{y_i})^2
\]

Measures unexplained variation (prediction errors).

---

## 3. Define R²

### Formula

:contentReference[oaicite:2]{index=2}

### Interpretation of R² = 0.85

The model explains 85% of the variation in the target variable.

---

## 4. Boundary Cases

### (a) Perfect Fit

All points lie exactly on the regression line.

- SSE = 0
- SSR = SST
- R² = 1

### (b) Predict Mean for Every Point

Model predicts only ȳ.

- SSR = 0
- SSE = SST
- R² = 0

---

## 5. Dataset Calculation

### Given

x = [1,2,3,4,5]

y = [2,4,5,4,5]

Regression Model:

ŷ = 2.2 + 0.6x

### Mean of y

ȳ = 4

---

### Predicted Values

| x | y | ŷ |
|---|---|---|
|1|2|2.8|
|2|4|3.4|
|3|5|4.0|
|4|4|4.6|
|5|5|5.2|

---

### SST Calculation

| y | y−ȳ | (y−ȳ)² |
|---|---|---|
|2|-2|4|
|4|0|0|
|5|1|1|
|4|0|0|
|5|1|1|

SST = 4 + 0 + 1 + 0 + 1

### SST = 6

---

### SSE Calculation

| y | ŷ | e = y−ŷ | e² |
|---|---|---|---|
|2|2.8|-0.8|0.64|
|4|3.4|0.6|0.36|
|5|4.0|1.0|1.00|
|4|4.6|-0.6|0.36|
|5|5.2|-0.2|0.04|

SSE = 0.64 + 0.36 + 1 + 0.36 + 0.04

### SSE = 2.40

---

### SSR Calculation

| ŷ | ŷ−ȳ | (ŷ−ȳ)² |
|---|---|---|
|2.8|-1.2|1.44|
|3.4|-0.6|0.36|
|4.0|0|0|
|4.6|0.6|0.36|
|5.2|1.2|1.44|

SSR = 1.44 + 0.36 + 0 + 0.36 + 1.44

### SSR = 3.60

---

### Verification

SST = SSR + SSE

6 = 3.60 + 2.40

6 = 6 ✅

Verified.

---

## 6. Compute R²

### Formula

R² = 1 − (SSE / SST)

Substitute values:

R² = 1 − (2.40 / 6)

R² = 1 − 0.40

### Final Answer

R² = 0.60

or

### R² = 60%

---

## Interpretation

60% of the variation in y is explained by x through the regression model.

The remaining 40% is due to unexplained factors and prediction errors.

---

# Final Summary

|----------|--------|
| SST | 6.00 |
| SSE | 2.40 |
| SSR | 3.60 |
| R² | 0.60 |
| Explained Variance | 60% |

SST = SSR + SSE

# QUEST 02 — Error Metrics: MAE, MSE & RMSE

---

## 1. Define MAE, MSE and RMSE

### MAE (Mean Absolute Error)

:contentReference[oaicite:0]{index=0}

Measures the average absolute prediction error.

### Unit

Same as y

---

### MSE (Mean Squared Error)

:contentReference[oaicite:1]{index=1}

Measures the average squared error.

### Unit

(y unit)²

---

### RMSE (Root Mean Squared Error)

:contentReference[oaicite:2]{index=2}

Measures the average prediction error while heavily penalizing large errors.

### Unit

Same as y

---

## 2. Compute MAE, MSE and RMSE

### Given

| y | ŷ |
|---|---|
|2|2.8|
|4|3.4|
|5|4.0|
|4|4.6|
|5|5.2|

---

### Residual Table

| y | ŷ | e = y−ŷ | \|e\| | e² |
|---|---|---|---|---|
|2|2.8|-0.8|0.8|0.64|
|4|3.4|0.6|0.6|0.36|
|5|4.0|1.0|1.0|1.00|
|4|4.6|-0.6|0.6|0.36|
|5|5.2|-0.2|0.2|0.04|

### Sums

Σ|e| = 3.2

Σe² = 2.4

---

### MAE

MAE = 3.2 / 5

### MAE = 0.64

---

### MSE

MSE = 2.4 / 5

### MSE = 0.48

---

### RMSE

RMSE = √0.48

### RMSE ≈ 0.693

---

## 3. Units Interpretation

Suppose y = House Price (Lakhs ₹)

| Metric | Unit |
|----------|----------|
| MAE | Lakhs ₹ |
| MSE | (Lakhs ₹)² |
| RMSE | Lakhs ₹ |

### Most Interpretable Metric

MAE

Because it directly tells the average prediction error in the same units as house price.

## 4. Sensitivity to Outliers

### Add Outlier

x = 6

y = 20

Prediction:

ŷ = 2.2 + 0.6(6)

ŷ = 5.8

Residual:

e = 20 − 5.8

e = 14.2

### New Totals

Σ|e| = 3.2 + 14.2

Σ|e| = 17.4

Σe² = 2.4 + (14.2)²

Σe² = 2.4 + 201.64

Σe² = 204.04

n = 6

---

### New MAE

MAE = 17.4 / 6

### MAE = 2.90

---

### New MSE

MSE = 204.04 / 6

### MSE = 34.01

---

### New RMSE

RMSE = √34.01

### RMSE ≈ 5.83

---

### Comparison

| Metric | Before | After Outlier |
|----------|----------|----------|
| MAE | 0.64 | 2.90 |
| RMSE | 0.69 | 5.83 |

### Which Changed More?

RMSE changed much more because squaring amplifies large errors.

---

## 5. Metric Selection

### Scenario

Large house-price prediction mistakes are very costly.

Example:

Overestimating by ₹20 Lakhs.

### Best Choice

RMSE

### Reason

RMSE penalizes large errors much more heavily than MAE.

This helps reduce costly mistakes.

---

## 6. R² vs RMSE

### Model A

R² = 0.92

RMSE = ₹8 Lakhs

---

### Model B

R² = 0.88

RMSE = ₹4 Lakhs

---

### Preferred Model

Model B

### Reason

Although its R² is slightly lower, its prediction errors are half as large.

In real business applications, prediction error often matters more than a small increase in R².

---

# Final Summary

|----------|----------|
| MAE | 0.64 |
| MSE | 0.48 |
| RMSE | 0.69 |

### Key Takeaway

- MAE treats all errors equally.
- RMSE penalizes large errors heavily.
- Use RMSE when large mistakes are expensive.
- Use MAE when interpretability is important.

# QUEST 03 — Hypothesis Test for the Slope β₁

---

## 1. State the Hypotheses

### Null Hypothesis (H₀)

β₁ = 0

There is no linear relationship between x and y.

### Alternative Hypothesis (H₁)

β₁ ≠ 0

There is a linear relationship between x and y.

### Interpretation

Rejecting H₀ means x significantly affects y and the slope is not zero.

---

## 2. Standard Error of β₁

Formula:

:contentReference[oaicite:0]{index=0}

where

:contentReference[oaicite:1]{index=1}

### Why n − 2?

Because two parameters are estimated:

- β₀ (Intercept)
- β₁ (Slope)

Therefore we lose 2 degrees of freedom.

---

## 3. Compute MSE, SE(β₁) and t-statistic

### Given

β₁ = 0.6

n = 5

SSE = 1.2

Σ(xᵢ − x̄)² = 10

---

### Step 1: Calculate MSE

MSE = SSE / (n − 2)

MSE = 1.2 / (5 − 2)

MSE = 1.2 / 3

### MSE = 0.4

---

### Step 2: Calculate SE(β₁)

SE(β₁) = √(0.4 / 10)

SE(β₁) = √0.04

### SE(β₁) = 0.2

---

### Step 3: Calculate t-statistic

t = β₁ / SE(β₁)

t = 0.6 / 0.2

### t = 3.0

---

## 4. Decision Rule

### Given

α = 0.05

Degrees of Freedom = n − 2 = 3

Critical Value:

t* = 3.182

---

### Compare

| Value | Result |
|---------|---------|
| Calculated t | 3.0 |
| Critical t | 3.182 |

Since

3.0 < 3.182

### Decision

Fail to Reject H₀

### Conclusion

There is not enough evidence at the 5% significance level to conclude that x and y are linearly related.

---

## 5. p-value Interpretation

### Given

p = 0.03

This means the observed result would be unusual if the null hypothesis were true.

Since

0.03 < 0.05

We reject H₀ at α = 0.05.

---

### If α = 0.01

Since

0.03 > 0.01

We fail to reject H₀.

### Conclusion

The result is significant at 5% but not significant at 1%.

---

## 6. Practical vs Statistical Significance

### Given

β₁ = 0.001

p = 0.0002

n = 50,000

---

### Interpretation

Although the result is statistically significant, the effect size is extremely small.

A change of 1 unit in x changes y by only 0.001 units, which may have no practical business value.

---

# Final Summary

| Quantity | Value |
|-----------|---------|
| β₁ | 0.6 |
| SSE | 1.2 |
| n | 5 |

### Key Idea

Statistical significance tells us whether an effect exists.

Practical significance tells us whether that effect is large enough to matter in the real world.

```text
      Confidence Band
          /\
         /  \
        /    \
-------/------\-------
      Regression Line
       \      /
        \    /
         \  /
          \/

Observed Data Range
|------------------|

Danger Zone
<--------------->

Outside observed range

Explanation

Confidence bands are narrow near x̄ and become wider near the edges.

Predictions beyond the observed range are called extrapolation and are highly unreliable.

**6. Practical Use**

**Use CI When**

A bank wants to estimate the average repayment amount for all customers with similar characteristics.

Example:
- "What is the average repayment for customers earning ₹50,000/month?"

Use PI When

A bank wants to predict the repayment amount of one specific customer.

Example:
- "How much will Rahul repay?"

**Risk of Choosing Wrong Interval**

Using CI instead of PI for an individual customer underestimates uncertainty and may lead to poor lending decisions.

| Quantity      | Value            |
| ------------- | ---------------- |
| β₁            | 0.6              |
| SE(β₁)        | 0.19             |
| 95% CI for β₁ | (-0.005 , 1.205) |
| SE_mean       | 0.283            |
| CI at x₀=3    | (3.10 , 4.90)    |
| SE_pred       | 0.693            |
| PI at x₀=3    | (1.80 , 6.21)    |


# Answers

## 1. What does SST measure — what would it mean if SST = 0?

SST (Total Sum of Squares) measures the total variation of y around its mean.

:contentReference[oaicite:0]{index=0}

If SST = 0, all y values are identical and there is no variation in the dataset.

---

## 2. State the variance decomposition identity SST = SSR + SSE and explain what each term represents.

:contentReference[oaicite:1]{index=1}

- SST = Total variation in y
- SSR = Variation explained by the regression model
- SSE = Variation not explained by the model (errors)

---

## 3. What is R² — write its formula and interpret a value of 0.72 in plain English.

Formula:

:contentReference[oaicite:2]{index=2}

Interpretation:

R² = 0.72 means the model explains 72% of the variation in y, while 28% remains unexplained.

---

## 4. What are the two extreme boundary values of R² — what does each mean geometrically?

### R² = 1
Perfect fit.

All points lie exactly on the regression line.

### R² = 0
No explanatory power.

The regression line predicts only the mean value ȳ.

---

## 5. Can R² be negative — under what circumstance, and what would it mean about the model?

Yes.

R² becomes negative when the model performs worse than simply predicting the mean of y.

It indicates a very poor model fit.

---

## 6. What is the difference between MAE and RMSE — which is more sensitive to outliers and why?

### MAE

:contentReference[oaicite:3]{index=3}

Measures average absolute error.

### RMSE

:contentReference[oaicite:4]{index=4}

Squares errors before averaging.

RMSE is more sensitive to outliers because large errors are squared and receive much larger penalties.

---

## 7. Why is RMSE measured in the same units as y but MSE is not?

MSE is measured in squared units because errors are squared.

RMSE takes the square root of MSE, bringing the units back to the original units of y.

### Practical Problem

MSE is harder to interpret because its units are squared.

---

## 8. State the null and alternative hypotheses for testing the slope β₁ in SLR.

### Null Hypothesis

H₀: β₁ = 0

(No linear relationship)

### Alternative Hypothesis

H₁: β₁ ≠ 0

(Linear relationship exists)

---

## 9. Why are n−2 degrees of freedom used in the t-test for β₁, not n or n−1?

Two parameters are estimated from the data:

- β₀ (Intercept)
- β₁ (Slope)

Therefore two degrees of freedom are lost.

Degrees of Freedom:

n − 2

---

## 10. What is the difference between statistical significance and practical significance?

### Statistical Significance

Shows whether an effect is likely real.

### Practical Significance

Shows whether the effect is large enough to matter.

### Example

β₁ = 0.001 with p = 0.0001

Statistically significant but practically useless because the effect is extremely small.

---

## 11. What is the difference between a confidence interval for the mean response and a prediction interval?

### Confidence Interval (CI)

Estimates where the average response lies.

### Prediction Interval (PI)

Estimates where an individual future observation lies.

### Which is Wider?

Prediction Interval is wider because it includes both:

- Model uncertainty
- Individual variability

---

## 12. Why does the confidence band for a regression line narrow near x̄ and wider near the extremes of x?

Most information is concentrated around the mean x̄.

Therefore predictions near x̄ are more reliable and have smaller uncertainty.

As we move toward the extremes, there are fewer nearby observations, causing uncertainty and confidence bands to widen.
