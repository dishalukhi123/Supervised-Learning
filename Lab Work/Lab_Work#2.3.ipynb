# Mean Absolute Error (MAE) – Deep Dive

#### Q1. Formula — Write MAE = (1/n)Σ|yi − ŷi|; define each symbol and state the units of MAE relative to the original y units.

## Formula

MAE = (1/n) × Σ|yi - ŷi|

Where:
- yi = Actual Value
- ŷi = Predicted Value
- |yi - ŷi| = Absolute Error
- n = Number of Observations

MAE measures the average prediction error.

Units:
MAE has the same unit as the target variable.

```python
actual = [3, 5, 2, 8, 6]
predicted = [2.5, 4.8, 2.1, 7.0, 6.5]

errors = []

for a, p in zip(actual, predicted):
    errors.append(abs(a - p))

mae = sum(errors) / len(errors)

print("MAE =", mae)
```

#### Q1. MSE Formula — Write MSE = (1/n)Σ(yi − ŷi)²; compute each ei² and then MSE.


Answer

Mean Squared Error (MSE) measures the average squared difference between actual and predicted values.




Dataset

Actual (y)     = [3, 5, 2, 8, 6]
Predicted (ŷ)  = [2.5, 4.8, 2.1, 7.0, 6.5]

Calculation Table

| Actual | Predicted | Residual (ei) | ei²  |
| ------ | --------- | ------------- | ---- |
| 3      | 2.5       | 0.5           | 0.25 |
| 5      | 4.8       | 0.2           | 0.04 |
| 2      | 2.1       | -0.1          | 0.01 |
| 8      | 7.0       | 1.0           | 1.00 |
| 6      | 6.5       | -0.5          | 0.25 |






```python
actual = [3,5,2,8,6]
predicted = [2.5,4.8,2.1,7.0,6.5]

squared_errors = []

for a,p in zip(actual,predicted):
    error = a-p
    squared_errors.append(error**2)

mse = sum(squared_errors)/len(squared_errors)

print("Squared Errors:", squared_errors)
print("MSE =", mse)
```

#### Q6. Limitation of R²
**Problem**
Adding new features can never decrease R².

- 5 useful features

- 50 features (45 are noise)

### QUEST 04 — Adjusted R² (Penalising Unnecessary Complexity)
#### Q1. Motivation — Why is R² not enough?
**Answer**

R² always increases or remains the same when new features are added, even if those features are useless.

Therefore, R² may favor a more complex model that does not actually improve predictive performance.


Q2. Formula — Adjusted R²
Formula
Adjusted R2=1−(1−R2)(n−1)/(n−k−1)
	​

Definitions

| Symbol | Meaning                      |
| ------ | ---------------------------- |
| R²     | Coefficient of Determination |
| n      | Number of observations       |
| k      | Number of predictors         |
| n−1    | Total degrees of freedom     |
| n−k−1  | Residual degrees of freedom  |



**Why does it penalize complexity?**

When:

k increases

the denominator:

(n−k−1)

becomes smaller.
This increases the penalty term and reduces Adjusted R² unless the new feature improves R² sufficiently.



#### Q3. Worked Example

Model A
R² = 0.82
n = 50
k = 3

Calculation
Adjusted R2=1−(1−0.82)(49)/46
	       =1−8.82/46
           =1−0.1917
Adjusted R2A=0.8083


Model B
R² = 0.84
n = 50
k = 10


Calculation
Adjusted R2=1−(1−0.84)(49)/39
	       =1−7.84/39
	       =1−0.2010
Adjusted R2B=0.7990

| Model | R²   | Adjusted R² |
| ----- | ---- | ----------- |
| A     | 0.82 | 0.808       |
| B     | 0.84 | 0.799       |


Best Model

✅ Model A

Although Model B has a higher R², its extra predictors do not justify the added complexity.


```python
r2_a = 0.82
n = 50
k_a = 3

adj_a = 1 - ((1-r2_a)*(n-1)/(n-k_a-1))

# Model B
r2_b = 0.84
k_b = 10

adj_b = 1 - ((1-r2_b)*(n-1)/(n-k_b-1))

print("Adjusted R² A =", round(adj_a,4))
print("Adjusted R² B =", round(adj_b,4))
```

#### Q6. Full Metric Comparison Table

| Metric      | Formula                 | Units     | Range  | Sensitive to Outliers? | Best Used When                             |
| ----------- | ----------------------- | --------- | ------ | ---------------------- | ------------------------------------------ |
| MAE         | (1/n)Σ|yi−ŷi|           | Same as y | 0 → ∞  | No                     | Equal importance to all errors             |
| MSE         | (1/n)Σ(yi−ŷi)²          | y²        | 0 → ∞  | Yes                    | Training optimization                      |
| RMSE        | √MSE                    | Same as y | 0 → ∞  | Yes                    | Reporting prediction error                 |
| R²          | 1−SSE/SST               | Unitless  | (-∞,1] | No                     | Goodness of fit                            |
| Adjusted R² | 1−((1−R²)(n−1)/(n−k−1)) | Unitless  | (-∞,1] | No                     | Comparing models with different predictors |


```python
r2_a = 0.82
r2_b = 0.84

n = 50
k_a = 3
k_b = 10

adj_a = 1 - ((1-r2_a)*(n-1)/(n-k_a-1))
adj_b = 1 - ((1-r2_b)*(n-1)/(n-k_b-1))

comparison = pd.DataFrame({
    "Model":["A","B"],
    "R²":[r2_a,r2_b],
    "Adjusted R²":[adj_a,adj_b]
})

print(comparison)

if adj_a > adj_b:
    print("\nBest Model: A")
else:
    print("\nBest Model: B")
```
