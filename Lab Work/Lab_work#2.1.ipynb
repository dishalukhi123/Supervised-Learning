# QUEST 01 — The Model Equation & Cost Function

---

## 1. Write the SLR Model and Define Each Symbol

### Simple Linear Regression Equation

ŷᵢ = β₀ + β₁xᵢ + εᵢ

| Symbol | Meaning |
|----------|----------|
| ŷᵢ | Predicted value of y |
| β₀ | Intercept (starting value) |
| β₁ | Slope (rate of change) |
| εᵢ | Error term (difference between actual and predicted value) |

---

## 2. Residual Definition

### Equation

eᵢ = yᵢ − ŷᵢ

### Interpretation

- Positive Residual (eᵢ > 0):
  
  Actual value is above the regression line.

- Negative Residual (eᵢ < 0):
  
  Actual value is below the regression line.

### Scatter Plot View

Above Line → Positive Residual

Below Line → Negative Residual

---

## 3. Cost Function (SSE)

### Sum of Squared Errors

SSE = Σeᵢ²

or

SSE = Σ(yᵢ − β₀ − β₁xᵢ)²

### Why Square Errors?

1. Prevents positive and negative errors from canceling out.
2. Gives larger errors a bigger penalty.

---

## 4. Mean Squared Error (MSE)

### Formula

MSE = (1/n) × SSE

### Advantage

Dividing by n makes error values comparable across datasets of different sizes.

---

## 5. Geometric Interpretation

Minimizing SSE means finding the line whose total squared vertical distances from all data points is the smallest.

The best-fit line is the only line that produces the minimum possible SSE for the dataset.

---

## 6. Why Not Absolute Error?

### Absolute Error

Σ|eᵢ|

### Mathematical Reason

Absolute value is not differentiable at zero, making optimization more difficult.

### Practical Reason

Squared errors penalize large mistakes more heavily, leading to better model fitting and a unique optimal solution.

---

# QUEST 02 — Setting Up the Partial Derivatives

---

## 1. Objective Function

The goal is to find the values of β₀ and β₁ that minimize the Sum of Squared Errors (SSE).

### Loss Function

L(β₀, β₁) = Σ(yᵢ − β₀ − β₁xᵢ)²

We want:

min L(β₀, β₁)

with respect to β₀ and β₁.

---

## 2. Partial Derivative with Respect to β₀

### Loss Function

L = Σ(yᵢ − β₀ − β₁xᵢ)²

### Apply Chain Rule

∂L/∂β₀

= Σ 2(yᵢ − β₀ − β₁xᵢ)(−1)

### Before Simplification

∂L/∂β₀

= −2Σ(yᵢ − β₀ − β₁xᵢ)

---

## 3. Set ∂L/∂β₀ = 0

−2Σ(yᵢ − β₀ − β₁xᵢ) = 0

Divide by −2

Σ(yᵢ − β₀ − β₁xᵢ) = 0

Expand

Σyᵢ − nβ₀ − β₁Σxᵢ = 0

Rearrange

### First Normal Equation

Σyᵢ = nβ₀ + β₁Σxᵢ

---

## 4. Partial Derivative with Respect to β₁

### Apply Chain Rule

∂L/∂β₁

= Σ 2(yᵢ − β₀ − β₁xᵢ)(−xᵢ)

### Before Simplification

∂L/∂β₁

= −2Σxᵢ(yᵢ − β₀ − β₁xᵢ)

---

## 5. Set ∂L/∂β₁ = 0

−2Σxᵢ(yᵢ − β₀ − β₁xᵢ) = 0

Divide by −2

Σxᵢ(yᵢ − β₀ − β₁xᵢ) = 0

Expand

Σxᵢyᵢ − β₀Σxᵢ − β₁Σxᵢ² = 0

Rearrange

### Second Normal Equation

Σxᵢyᵢ = β₀Σxᵢ + β₁Σxᵢ²

---

## 6. System of Equations

### Equation 1 (from ∂L/∂β₀)

Σyᵢ = nβ₀ + β₁Σxᵢ

### Equation 2 (from ∂L/∂β₁)

Σxᵢyᵢ = β₀Σxᵢ + β₁Σxᵢ²

---

## Matrix Form

| Equation Source | Normal Equation |
|----------------|-----------------|
| ∂L/∂β₀ = 0 | Σyᵢ = nβ₀ + β₁Σxᵢ |
| ∂L/∂β₁ = 0 | Σxᵢyᵢ = β₀Σxᵢ + β₁Σxᵢ² |

---

## Final Result

By solving these two normal equations simultaneously, we obtain:

- β₀ (Intercept)
- β₁ (Slope)

which define the Ordinary Least Squares (OLS) best-fit regression line.

# QUEST 03 — Solving for β₀ and β₁ (Closed-Form Solution)

---

## 1. Divide Normal Equation 1 by n

### First Normal Equation

Σyᵢ = nβ₀ + β₁Σxᵢ

Divide both sides by n:

(Σyᵢ)/n = β₀ + β₁(Σxᵢ)/n

Using:

x̄ = (Σxᵢ)/n

ȳ = (Σyᵢ)/n

Therefore:

### β₀ = ȳ − β₁x̄

---

## 2. Substitute into Normal Equation 2

### Second Normal Equation

Σxᵢyᵢ = β₀Σxᵢ + β₁Σxᵢ²

Substitute:

β₀ = ȳ − β₁x̄

Then:

Σxᵢyᵢ = (ȳ − β₁x̄)Σxᵢ + β₁Σxᵢ²

Expand:

Σxᵢyᵢ = ȳΣxᵢ − β₁x̄Σxᵢ + β₁Σxᵢ²

Move terms:

Σxᵢyᵢ − ȳΣxᵢ = β₁(Σxᵢ² − x̄Σxᵢ)

---

## 3. Solve for β₁

Since:

Σxᵢ = nx̄

Substitute:

Σxᵢyᵢ − nx̄ȳ = β₁(Σxᵢ² − nx̄²)

Therefore:

β₁ = (Σxᵢyᵢ − nx̄ȳ) / (Σxᵢ² − nx̄²)

Rearranging gives the standard formula:

### β₁ = Σ(xᵢ − x̄)(yᵢ − ȳ) / Σ(xᵢ − x̄)²

---

## 4. Recover β₀

Substitute β₁ into:

β₀ = ȳ − β₁x̄

### Final Formula

β₀ = ȳ − x̄ × [ Σ(xᵢ − x̄)(yᵢ − ȳ) / Σ(xᵢ − x̄)² ]

---

## 5. Interpretation of β₁

### Ratio Form

β₁ = Cov(x,y) / Var(x)

Where:

- Cov(x,y) = Covariance between x and y
- Var(x) = Variance of x

### Meaning

β₁ measures how much y changes when x changes.

### If β₁ = 0

A slope of zero means there is no linear relationship between x and y.

The regression line becomes horizontal.

---

## 6. Dimensional Consistency

Suppose:

- x = Years
- y = Rupees (₹)

### Unit of β₁

β₁ = Rupees / Year

(₹ per year)

### Unit of β₀

β₀ = Rupees

(₹)

### Check Equation

ŷ = β₀ + β₁x

Units:

₹ = ₹ + (₹/year × year)

₹ = ₹ + ₹

✓ Units match correctly.

Therefore, the regression equation is dimensionally consistent.

---

# Final Closed-Form Solution

### Slope (β₁)

β₁ = Σ(xᵢ − x̄)(yᵢ − ȳ) / Σ(xᵢ − x̄)²

### Intercept (β₀)

β₀ = ȳ − β₁x̄

### Regression Line

ŷ = β₀ + β₁x

# QUEST 04 — Full Worked Derivation on a Real Dataset

---

## 1. Given Dataset

x = [1, 2, 3, 4, 5]

y = [2, 4, 5, 4, 5]

### Number of Observations

n = 5

### Mean of x

x̄ = (1 + 2 + 3 + 4 + 5) / 5

x̄ = 15 / 5

x̄ = 3

### Mean of y

ȳ = (2 + 4 + 5 + 4 + 5) / 5

ȳ = 20 / 5

ȳ = 4

---

## 2. Computation Table

| xᵢ | yᵢ | (xᵢ−x̄) | (yᵢ−ȳ) | (xᵢ−x̄)(yᵢ−ȳ) | (xᵢ−x̄)² |
|----|----|---------|---------|----------------|----------|
| 1 | 2 | -2 | -2 | 4 | 4 |
| 2 | 4 | -1 | 0 | 0 | 1 |
| 3 | 5 | 0 | 1 | 0 | 0 |
| 4 | 4 | 1 | 0 | 0 | 1 |
| 5 | 5 | 2 | 1 | 2 | 4 |

### Column Sums

Σ(xᵢ−x̄)(yᵢ−ȳ) = 6

Σ(xᵢ−x̄)² = 10

---

## 3. Compute β₁

### Formula

β₁ = Σ(xᵢ−x̄)(yᵢ−ȳ) / Σ(xᵢ−x̄)²

Substitute values:

β₁ = 6 / 10

### Answer

β₁ = 0.6

---

## 4. Compute β₀

### Formula

β₀ = ȳ − β₁x̄

Substitute values:

β₀ = 4 − (0.6 × 3)

β₀ = 4 − 1.8

### Answer

β₀ = 2.2

---

## 5. Final Regression Model

### Equation

ŷ = β₀ + β₁x

Substitute values:

ŷ = 2.2 + 0.6x

### Predict for x = 6

ŷ = 2.2 + (0.6 × 6)

ŷ = 2.2 + 3.6

### Prediction

ŷ = 5.8

---

## 6. Compute SSE

### Predicted Values

| x | y | ŷ = 2.2 + 0.6x |
|---|---|---|
| 1 | 2 | 2.8 |
| 2 | 4 | 3.4 |
| 3 | 5 | 4.0 |
| 4 | 4 | 4.6 |
| 5 | 5 | 5.2 |

---

### Residual Table

| y | ŷ | e = y−ŷ | e² |

| 2 | 2.8 | -0.8 | 0.64 |
| 4 | 3.4 | 0.6 | 0.36 |
| 5 | 4.0 | 1.0 | 1.00 |

### SSE Calculation

SSE = Σe²

## 2. What is a residual — how is it different from an error term εᵢ in the population model?

### Residual

eᵢ = yᵢ − ŷᵢ

Residual is the difference between the actual value and the predicted value in a sample.

### Error Term

εᵢ = yᵢ − E(yᵢ)

The error term represents unknown population error.

**Difference:** Residuals are observable; error terms are theoretical and unobservable.

---

## 3. Why does OLS minimise the sum of squared errors rather than the sum of absolute errors?

OLS uses squared errors because:

1. Squaring prevents positive and negative errors from canceling.
2. Squared functions are differentiable, making calculus optimization easier.
3. Large errors receive a larger penalty.
