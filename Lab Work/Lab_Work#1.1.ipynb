**02. Problem Definition — Loan Default Prediction**

| Component             | Answer                                                        |
| --------------------- | ------------------------------------------------------------- |
| Business Goal         | Predict whether a customer will default on a loan             |
| ML Task Type          | Classification                                                |
| Success Metric        | F1 Score / AUC Score                                          |
| Failure in Production | Wrongly approving risky customers or rejecting good customers |

**03. Data Collection & Preprocessing**

- Data Sources
1. Customer bank records
2. Credit score history
3. Loan repayment history

**Common Data Quality Issues**

| Problem                | Solution               |
| ---------------------- | ---------------------- |
| Missing Values         | Fill using mean/median |
| Duplicate Records      | Remove duplicates      |
| Incorrect Data Formats | Standardize formats    |


**04. EDA Phase Questions**

1. Is there any missing or duplicate data?
2. Which features are strongly related to loan default?
3. Is the dataset balanced or imbalanced?


**05. Model Evaluation**

Accuracy alone is poor for loan default prediction because the dataset may be imbalanced.

Example:
- 95% customers repay loans
- 5% customers default

A model predicting all customers as “non-default” can still achieve 95% accuracy.

**Better Metric**

F1 Score or AUC Score

because they balance:

- precision
- recall
- classification performance

for minority default cases.

**06. Deployment & Monitoring
✅ Model Drift

Model drift happens when real-world data changes over time and model performance decreases.


**Example Sign of Drift**

- Sudden increase in wrong loan default predictions
- Drop in F1 Score or accuracy after deployment

### QUEST 04 — Data Science Job Roles

**01. Comparison Table**

| Category           | Data Engineer          | Data Analyst              | Data Scientist          | ML Engineer             |
| ------------------ | ---------------------- | ------------------------- | ----------------------- | ----------------------- |
| Primary Goal       | Build data pipelines   | Analyze business data     | Build predictive models | Deploy ML models        |

**02. Data Engineer**

A Data Engineer builds systems and pipelines that collect, clean, and store large amounts of data.
Without clean and reliable data pipelines, Data Analysts and Data Scientists cannot perform analysis or build ML models properly.

**03. Data Analyst vs Data Scientist**

A Data Analyst focuses on dashboards, reports, and business insights, while a Data Scientist builds machine learning models and prediction systems.

Example:
- Data Analyst → Monthly sales dashboard
- Data Scientist → Customer churn prediction model


**04. Data Scientist vs ML Engineer**

Before deployment, an ML Engineer must:

1. Deploy the model into production using APIs or cloud services
2. Monitor model performance, scalability, and reliability

**05. Role Overlap & Handoff**

The Data Engineer’s work ends after creating clean, structured, and accessible datasets.
The Data Scientist’s work begins when the prepared data is used for analysis, feature engineering, and model building.

# Machine Learning Fundamentals — Answers

---

## 1. What is Machine Learning?

Machine Learning is a way of teaching systems to improve their performance through experience instead of manually giving every instruction.

---

## 2. Difference Between Machine Learning and Traditional Programming

Traditional programming follows fixed rules written by humans, while Machine Learning learns patterns from examples and makes decisions automatically. The fundamental shift is moving from manually coded logic to learning from experience.

---

## 3. Relationship Between AI, Machine Learning, and Deep Learning

Artificial Intelligence (AI) is the broad field of creating intelligent systems. Machine Learning (ML) is a subset of AI that enables systems to learn from data. Deep Learning (DL) is a subset of ML that uses neural networks with many layers to solve complex problems.

```text
AI
 └── Machine Learning
      └── Deep Learning
```

---

## 4. Three Main Types of Machine Learning

| Type                   | Description                        | Real-World Example    |
| ---------------------- | ---------------------------------- | --------------------- |
| Supervised Learning    | Learns using labeled data          | Email Spam Detection  |
| Unsupervised Learning  | Finds hidden patterns in data      | Customer Segmentation |
| Reinforcement Learning | Learns using rewards and penalties | Self-Driving Cars     |

---

## 5. Difference Between Label and Feature

A feature is an input variable used for prediction, while a label is the final output or target the model tries to predict.

### Example:

* Features → Study Hours, Attendance
* Label → Pass or Fail

---

## 6. Why Unsupervised Learning is Useful

Unsupervised learning can discover hidden patterns, groups, and relationships in data without needing labeled examples. It helps identify structures humans may not already know exist.

---

## 7. Role of Reward Signal in Reinforcement Learning

The reward signal tells the agent whether its action was good or bad. Without rewards, the agent cannot learn which actions improve performance or achieve goals.

---

## 8. Seven Phases of the Machine Learning Development Life Cycle

1. Problem Definition
2. Data Collection
3. Data Preprocessing
4. Exploratory Data Analysis (EDA)
5. Model Building
6. Model Evaluation
7. Deployment & Monitoring

---

## 9. Why Problem Definition is the Most Critical Phase

If the problem is defined incorrectly, the entire project can fail even if the model performs well technically. Poor problem definition leads to wrong objectives, incorrect data collection, and meaningless predictions.

---

## 10. What is Model Drift?

Model drift occurs when real-world data changes over time, causing model performance to decrease. Continuous monitoring is required to detect accuracy drops and retrain the model when needed.

---

## 11. Difference Between Data Scientist and ML Engineer

A Data Scientist focuses on analyzing data and building machine learning models, while an ML Engineer focuses on deploying, scaling, and maintaining those models in production systems.

---

## 12. Role of a Data Engineer

A Data Engineer builds data pipelines, storage systems, and infrastructure for collecting and processing data. Their role is foundational because all other data roles depend on reliable, clean, and accessible data.
