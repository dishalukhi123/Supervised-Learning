### Step 1: Mean

[
\bar{x}=\frac{2+4+6+8}{4}=5
]

[
\bar{y}=\frac{3+7+9+13}{4}=8
]

---

### Step 2: Calculate β₁

[
\beta_1=
\frac{\sum(x_i-\bar{x})(y_i-\bar{y})}
{\sum(x_i-\bar{x})^2}
]

[
\beta_1=\frac{32}{20}=1.6
]

---

### Step 3: Calculate β₀

[
\beta_0=\bar{y}-\beta_1\bar{x}
]

[
\beta_0=8-(1.6\times5)
]

[
\beta_0=0
]

---

### Final Equation

\hat{y}=1.6x

### QUEST 04 — Assumptions of Linear Regression

**01. Five Assumptions of Linear Regression**

| Assumption             | Explanation                                                                   |
| ---------------------- | ----------------------------------------------------------------------------- |
| Linearity              | The relationship between input features and target variable should be linear. |
| Independence of Errors | Prediction errors should be independent of each other.                        |
| Homoscedasticity       | Residuals should have constant variance across all predictions.               |
| Normality of Residuals | Residuals should follow a normal distribution.                                |
| No Multicollinearity   | Independent variables should not be highly correlated with each other.        |

**05. Multicollinearity**
Problem

If two features have a very high correlation (r = 0.97), the model cannot determine which feature is actually influencing the prediction.

This causes:

- Unstable β coefficients
- Difficult interpretation
- Increased variance

Detection Method

Variance Inflation Factor (VIF)

Fix Method
- Remove one correlated feature
- Apply PCA
- Use Ridge Regression

**06. Consequences of Violating Assumptions**

| Assumption Violated    | Consequence                                   |
| ---------------------- | --------------------------------------------- |
| Linearity              | Biased predictions                            |
| Independence of Errors | Incorrect statistical tests                   |
| Homoscedasticity       | Unreliable confidence intervals               |
| Normality of Residuals | Inaccurate hypothesis testing                 |
| Multicollinearity      | Unstable coefficients and poor interpretation |

| Normality         | Q-Q Plot                 |
| Multicollinearity | Correlation Matrix / VIF |

## 3. What is the fundamental difference between a regression problem and a classification problem?

**Answer:**

Regression predicts numerical values, whereas classification predicts categories or classes.

**Example:**

- Classification: Email Spam Detection

## 4. Why is Logistic Regression considered a classification algorithm despite having 'regression' in its name?

**Answer:**

Logistic Regression calculates the probability of belonging to a class and then assigns a category. Therefore, it is used for classification problems.

## 5. What evaluation metric would you use for a regression model — and a different one for a classification model — and why can't they be swapped?

**Answer:**

- Regression Metric: RMSE
- Classification Metric: F1 Score

## 6. Write the equation for Simple Linear Regression and explain what β₀ and β₁ represent in plain terms.

### Equation

ŷ = β₀ + β₁x

### Explanation

- β₀: Starting value when x = 0
- β₁: Change in y for every 1-unit increase in x

## 7. What does 'best-fit line' mean — best fit by what mathematical criterion?

**Answer:**

The best-fit line is the line that minimizes the Sum of Squared Errors (SSE) between actual and predicted values.

---

## 8. What is a residual in Linear Regression — what would it mean if all residuals were exactly 0?

**Answer:**

A residual is the difference between the actual value and the predicted value.

Residual = Actual Value − Predicted Value

If all residuals are 0, the model predicts every observation perfectly.

---

## 9. What is the Linearity assumption in Linear Regression — how do you test whether it holds?

**Answer:**

The linearity assumption states that the relationship between the feature and target should be linear.

**Test:** Scatter Plot

- Straight-line trend → Assumption satisfied
- Curved trend → Assumption violated

---

## 10. What is Homoscedasticity — what does a residual plot look like when this assumption is violated?

**Answer:**

Homoscedasticity means residuals have constant variance.

When violated, the residual plot shows a funnel-shaped pattern where error spread increases or decreases.

---

## 11. What is Multicollinearity — why does it make the β coefficients unreliable?

**Answer:**

Multicollinearity occurs when independent variables are highly correlated.

It makes β coefficients unstable because the model cannot clearly determine each feature's individual effect.

---

## 12. What is the Normality of residuals assumption — which diagnostic plot is used to check it, and what does the plot show?

**Answer:**

The normality assumption states that residuals should follow a normal distribution.

**Diagnostic Plot:** Q-Q Plot

- Points follow the diagonal line → Residuals are normal
- Points deviate from the line → Residuals are not normal
